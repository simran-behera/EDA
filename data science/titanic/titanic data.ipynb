```python
df3.groupby('Sex')['Age'].agg(['mean','median'])
```

* average age of male and female passengers are 30 and 27 which is very close to overall avg value.so no need to fill using sex avg values

```python
df3.groupby('Pclass')['Age'].agg(['mean','median'])
```

* average age of class1 is 38 , class2 is 29 , class3 is 25 which far from overall avg value.sowe need to fill using pclass avg values

```python
def fix_age(pclass):
    if pclass==1:
        return 38
    elif pclass==2:
        return 29
    else:
        return 25
```

* divide the data into 2 parts
    * dfa0 ->data samples with age values missing
    * dfa1 ->data samples with age values are not missing
    

```python
dfa0=df[df['Age'].isnull()]
```

```python
dfa0
```

```python
dfa1=df[~df['Age'].isnull()]
```

```python
dfa1
```

* apply()- inbuilt panda function to apply user defined function

```python
dfa0['Age']=dfa0['Pclass'].apply(fix_age)
```

```python
dfa0
```

* combine dfa0 and dfa1

```python
df4=pd.concat((dfa0,dfa1),ignore_index=True)
```

```python
df4
```
