```python
import pandas as pd
df=pd.read_csv("D:\data science\Walmart.csv")
df
```

```python
print(df['Order Date'].max())
print(df['Order Date'].min())
```

```python
#to convert a column with dates into date time format
df['Order Date']=pd.to_datetime(df['Order Date'],format='%d-%m-%Y')
df['Ship Date']=pd.to_datetime(df['Ship Date'],format='%d-%m-%Y')
```

```python
df.dtypes
```

```python
print(df['Order Date'].max())
print(df['Order Date'].min())
```

```python
df['O_date']=df['Order Date'].dt.day
df['O_month']=df['Order Date'].dt.month
df['O_year']=df['Order Date'].dt.year
```

```python
df.head()
```

```python
def quarter(month):
    if month in [1,2,3]:
        return 'Q1'
    elif month in [4,5,6]:
        return 'Q2'
    elif month in [7,8,9]:
        return 'Q3'
    else:
        return 'Q4'
```

```python
df['O_quarter']=df['O_month'].apply(quarter)
```

```python
df
```

calculate the duration company took to ship a product

```python
df['Ship Date']-df['Order Date']
```

```python
df['processing_time']=(df['Ship Date']-df['Order Date']).apply(lambda x: x.days)
```

```python
import numpy as np
table = pd.pivot_table(data=df,
              values=['Order ID','Quantity','Sales','Profit'],
              index='Category',
              aggfunc={
                  'Order ID' : pd.Series.nunique,
                  'Quantity' :np.sum,
                   'Sales':np.mean,
                   'Profit':np.mean
              })
table
```

```python
table=table.rename(columns={'Order ID' : 'Total Orders',
                  'Quantity' : 'Total Quantity',
                   'Sales': 'Average Sales',
                   'Profit': 'Average Profit'
})
table
```

```python
table=table.reindex(columns=[ 'Total Orders','Total Quantity',
                             'Average Sales','Average Profit'])
table
```

```python
table =table.sort_values(by='Total Quantity',ascending=False)
table
```

```python
table[['Average Sales','Average Profit']].plot.bar()
```

```python
table[['Average Sales','Average Profit']].plot.barh()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,12))
x_axis=np.arange(len(table.index))
plt.bar(x_axis-0.2,height=table['Average Sales'],width=0.4,label='Average Sales')
plt.bar(x_axis+0.2,height=table['Average Profit'],width=0.4,label='Average Profit')
plt.xticks(x_axis,table.index,rotation=45)
plt.legend()
plt.show()
```
