```python
df_item_price_quantity=df.groupby('item_name')['item_price','quantity'].agg(['sum','count'])
```

```python
df_item_price_quantity
```

```python
df_item_price_quantity.columns=['item_price_sum','item_count','item_quantity_sum','item_count']
```

```python
df_item_price_quantity
```

```python
df_item_price_quantity=df_item_price_quantity.iloc[:,0:3]
```

```python
df_item_price_quantity
```

```python
df_item_price_quantity['avg_order_price']=df_item_price_quantity['item_price_sum']/df_item_price_quantity['item_count']
df_item_price_quantity['avg_quantity_price']=df_item_price_quantity['item_price_sum']/df_item_price_quantity['item_quantity_sum']
```

```python
df_item_price_quantity
```

```python
df_item_price_quantity.plot.bar(y=['avg_order_price','avg_quantity_price'],figsize=(12,7))
```
